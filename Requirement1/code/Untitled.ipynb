```python
import pandas as pd
```

```python
product='pacifier'#hair_dryer microwave pacifier
train=pd.read_csv(product+'.tsv', sep='\t', header=0)
```

```python
# Âà†Èô§Êó†ÂÖ≥Â≠óÊÆµ
train=train.drop(columns=['marketplace','customer_id','review_id','product_id','product_category'])
train['product_title'] = train['product_title'].str.lower()
# Â∞Ü‚Äún","y"ËΩ¨Êç¢Êàê‚Äù0","1"
train.replace('n','0',inplace=True)
train.replace('y','1',inplace=True)
train.replace('N','0',inplace=True)
train.replace('Y','1',inplace=True)
# Â∞ÜÂ≠óÁ¨¶‰∏≤Á±ªÂûãÁöÑÊï∞ÊçÆËΩ¨Êç¢Êàêfloat
train[['star_rating','helpful_votes','total_votes','vine','verified_purchase']]\
    =train[['star_rating','helpful_votes','total_votes','vine','verified_purchase']].astype('float')
```

```python
train['product_parent']=train['product_parent'].astype('object')
```

```python
# Ê®°Á≥äÂåπÈÖç Ê∏ÖÈô§Ë°®‰∏≠Êó†ÂÖ≥ÁöÑ‰∫ßÂìÅ 
train=train[train['product_title'].str.contains(product)]
# Êó•ÊúüËΩ¨Êç¢
train.loc[:,'review_date']=pd.to_datetime(train.loc[:,'review_date'],format='%m/%d/%Y', errors='coerce')
# Âà†Èô§Â∏¶ÊúâÁ©∫ÂÄºÁöÑË°åÔºåhow='any'Âè™Ë¶ÅÊúâ‰∏Ä‰∏™Á©∫ÂÄºÔºåÂ∞±Âà†Èô§Êï¥Ë°å
'''
ÈªòËÆ§axis=0,how=‚Äòany‚Äô,Âà†Èô§Â∏¶ÊúâÁ©∫ÂÄºÁöÑË°åÔºåÂè™Ë¶ÅÊúâ‰∏Ä‰∏™Á©∫ÂÄºÔºåÂ∞±Âà†Èô§Êï¥Ë°å

axis=1,Âà†Èô§Â∏¶ÊúâÁ©∫ÂÄºÁöÑÂàóÔºåÂè™Ë¶ÅÊúâ‰∏Ä‰∏™Á©∫ÂÄºÔºåÂ∞±Âà†Èô§Êï¥Âàó

how=‚Äòall‚ÄôÔºå Êï¥Ë°åÈÉΩÊòØÁ©∫ÂÄºÊó∂ÔºåÊâç‰ºöË¢´Âà†Èô§

threshÔºåËÆæÂÆöÈòàÂÄºÔºåÁº∫Â§±ÂÄº‰∏™Êï∞Â§ß‰∫éËØ•ÈòàÂÄºÔºåÊï¥Ë°å(axis=0)ÊàñÊï¥Âàó(axis=1)Êâç‰ºöË¢´Âà†Èô§

subset=[1,2]Ôºå Âà†Èô§ÊåáÂÆöÂàóÔºà1,2Ôºâ‰∏≠ÂåÖÂê´Áº∫Â§±ÂÄºÁöÑË°å
'''
train=train.dropna(subset=train.columns,how='any')
```

```python
df1=train.groupby(["product_parent"],sort=True).size().reset_index(name='total_rates_count')
```

```python
df2=train.groupby(["product_parent",'star_rating'],sort=True).size().reset_index(name='stars_rate_count')
```

```python
import pandas as pd
product='pacifier'#hair_dryer microwave pacifier
df = pd.read_csv('cleaned_'+product+'.csv', header=0,index_col=0)
```

## ÂéªÈô§Êó†ÂÖ≥ÁöÑÊ†áÁÇπÁ¨¶Âè∑ÔºåÂ∞ÜÂ≠óÊØçÁªü‰∏ÄËΩ¨Êç¢‰∏∫Â∞èÂÜô

```python
# Â∞ÜÊ†áÁÇπÁ¨¶Âè∑Áî®Á©∫Ê†ºÊõøÊç¢
import re
cleanup_re = re.compile('[^a-z]+')# ‰∏çÂåÖÂê´ Â∞èÂÜôÂ≠óÊØçÁöÑbaiÂ≠óÁ¨¶ÂåπÈÖç
```

```python
def cleanup(sentence):
    sentence = str(sentence)
    sentence = sentence.lower()
    sentence = cleanup_re.sub(' ', sentence).strip()# strip()Âà†Èô§Â≠óÁ¨¶‰∏≤Â§¥Â∞æÁöÑÂ≠óÁ¨¶
    #sentence = " ".join(nltk.word_tokenize(sentence))
    return sentence
```

```python
senti.head()
```

```python
senti["Summary_Clean"] = senti["review_headline_body"].apply(cleanup)
senti.head()
```

```python
prediction =  {} ## For storing results of different classifiers
#Fitting Logistic Regression
from sklearn import linear_model
# C Ê≠£ÂàôÂåñÁ≥ªÊï∞ÁöÑÂÄíÊï∞
logreg = linear_model.LogisticRegression(solver='liblinear' , C=100)
logistic = logreg.fit(X_train_tfidf, train["senti"])
prediction['LogisticRegression'] = logreg.predict_proba(X_test_tfidf)[:,1]
print("Logistic Regression Accuracy : {}".format(logreg.score(X_test_tfidf , test["senti"])))
```

```python
#Fitting Multinomial NB
from sklearn.naive_bayes import MultinomialNB
model1 = MultinomialNB().fit(X_train_tfidf , train["senti"])
prediction['Multinomial'] = model1.predict_proba(X_test_tfidf)[:,1]
print("Multinomial Accuracy : {}".format(model1.score(X_test_tfidf , test["senti"])))
```

```python
#Fitting Bernouli NB
from sklearn.naive_bayes import BernoulliNB
model2 = BernoulliNB().fit(X_train_tfidf,train["senti"])
prediction['Bernoulli'] = model2.predict_proba(X_test_tfidf)[:,1]
print("Bernoulli Accuracy : {}".format(model2.score(X_test_tfidf , test["senti"])))
```

```python
from sklearn.svm import SVC
model3 = SVC(C=100,probability=True)
model3.fit(X_train_tfidf,train["senti"])
prediction['SVM'] = model3.predict_proba(X_test_tfidf)[:,1]
print("SVM Accuracy : {}".format(model3.score(X_test_tfidf , test["senti"])))
```

```python
model4 = linear_model.SGDClassifier(loss="modified_huber",penalty="L2").fit(X_train_tfidf,train["senti"])
prediction['SGDClassifier'] = model4.predict_proba(X_test_tfidf)[:,1]
print("SGD Classifier Accuracy : {}".format(model4.score(X_test_tfidf , test["senti"])))
```

## AUCÊõ¥Â§ßÁöÑÂàÜÁ±ªÂô®ÊïàÊûúÊõ¥Â•Ω

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
import numpy as np

#To check out which classifier is doing what
def formatt(x):
    if x == 'neg':
        return 0
    if x == 0:
        return 0
    return 1
vfunc = np.vectorize(formatt)

cmp = 0
colors = ['k', 'b', 'g', 'm', 'y']
for model, predicted in prediction.items():
    if model not in 'Naive':
        false_positive_rate, true_positive_rate, thresholds = roc_curve(test["senti"].map(vfunc), predicted)
        roc_auc = auc(false_positive_rate, true_positive_rate)
        plt.plot(false_positive_rate, true_positive_rate, colors[cmp], label='%s: AUC %0.2f'% (model,roc_auc))
        cmp += 1

plt.title('Classifiers comparaison with ROC')
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
fig=plt.gcf()
fig.savefig('ROC_'+product+'_all_withSW.png')
plt.show()
```

```python
# ÂØπÊØî‰∏âÁßçÊ®°ÂûãÁöÑÁ≤æÁ°ÆÁéá„ÄÅÂè¨ÂõûÁéá„ÄÅF1ÔºåÂä†ÊùÉÂêÑÈ°πÂæóÂàÜÔºåÂèëÁé∞SVMÊïàÊûúÊúÄÂ•Ω
test.senti = test.senti.replace(["pos" , "neg"] , [True , False] )
keys = prediction.keys()
for key in ['LogisticRegression','SVM','SGDClassifier']:
    print(" {}:".format(key))
    print(metrics.classification_report(test["senti"], prediction.get(key)>.5, 
                                        target_names = ["positive", "negative"]))
    print("\n")
```
